```python
import datetime
import pandas as pd
import pandas_datareader as web
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

start = datetime.datetime(2018, 1, 1)
end = datetime.datetime.now()
df = web.DataReader("AMZN", 'morningstar', start, end)
df.reset_index(inplace=True)
df.set_index("Date", inplace=True)
df = df.drop("Symbol", axis=1)
 
df.describe()
```

```python
df
```

```python
df_close = pd.DataFrame(df.Close)
```

```python
df_close
```

```python
df_close["MA_9"] = df_close.Close.rolling(9).mean().shift()
df_close["MA_21"] = df_close.Close.rolling(21).mean()
```

```python
plt.figure(figsize=(15,5))
plt.grid(True)
plt.suptitle("AMZN STOCKS",fontsize = 18)
plt.plot(df_close['Close'], label = "AMZN")
plt.plot(df_close['MA_9'], label = "MA 9 day")
#plt.plot(df_close['MA_21'], label = "MA 21 day")
plt.xlabel("Date", fontsize = 16)
plt.ylabel("Price",fontsize = 16)
plt.legend(loc = 2)
plt.show()
```

# Volatibilidad

```python
df_close["change"] = np.log(df_close["Close"] / df_close["Close"].shift())
```

```python
plt.plot(df_close.change)
plt.xlabel("Date", fontsize = 18)
plt.show()
```

```python
df_close["Volatility"] = df_close.change.rolling(21).std().shift()
df_close["Volatility"].plot()
plt.ylabel("Volatilidad")
plt.show()
```

```python
df_close["actual_change"] = df_close["Close"] - df_close["Close"] .shift(1)
```

```python
df_close['exp_change'] = df_close["Close"] .shift(1) * df_close["Volatility"]
```

```python
df_close
```

```python
df_close["Magnitude"] = df_close["actual_change"] / df_close["exp_change"]
```

```python
df_close["Magnitude"].plot()
plt.show()
```

```python
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
x = a
y = df_close["MA_9"]
e_ms = []

def obj(b, x, y, n):
    return np.sum((y - np.array([x**i for i in range(n + 1)]).T.dot(b))**2)
plt.figure(figsize=(10,10))
#plt.scatter(x, y, label = 'Datos', c = "gray")
plt.plot(x,df_close['Close'], label = "AMZN")
plt.plot(x,df_close['MA_9'], label = "MA 9 day")

for i in range(2,3):
        
    b0 = np.random.random((i + 2,))
    res = opt.minimize(obj, b0, args=(x,y,i + 1))
    yhat = np.array([x**j for j in range(i + 2)]).T.dot(res.x)
    e_ms.append(sum((y - yhat)**2))
    plt.plot(x,np.array([x**j for j in range(i + 2)]).T.dot(res.x), label = "Grado %s"%(i+1))
menor = min(e_ms)
grado = e_ms.index(menor)+1

print(e_ms)
print(grado)
plt.legend(bbox_to_anchor= [1,.92])
plt.show()
```
