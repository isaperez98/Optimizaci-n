# Forma numérica

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib as mpl
label_size = 14
mpl.rcParams['xtick.labelsize']  = label_size
mpl.rcParams['ytick.labelsize']  = label_size
```

```python
r=1
def poblacion(x, t):
    return r * x *(1-x)
```

```python
x0 = 0.05
tt = np.linspace(0,10)

xx=odeint(poblacion, x0, tt)

plt.plot(tt, xx)
plt.xlabel('$t$',fontsize = 18)
plt.ylabel('$x$',fontsize = 18)
plt.show()
```

# Forma Analítica

```python
def logi_sol(t, x0, r):
    return 1/(1 + (1/x0 -1) * np.exp(-r * t))
```

```python
t = np.linspace(0,10)
for r in np.arange(-1, 1, 0.1):
    plt.plot(t, logi_sol(t, 0.05, r))
plt.xlabel('$t$', fontsize = 18)
plt.ylabel('$x$', fontsize = 18)
plt.show()
```

# Mapa Logístico

```python
import numpy as np
def mapa_logistico(r,x):
    return r* x * (1-x)
```

```python
r=3.991
#r=3.1
#r=3.5
#r=1
n=100
x= np.zeros(n)
x[0]=0.3

for i in range (n-1):
    x[i+1]= mapa_logistico(r,x[i])
    
plt.plot(np.arange(0,n),x)
```
